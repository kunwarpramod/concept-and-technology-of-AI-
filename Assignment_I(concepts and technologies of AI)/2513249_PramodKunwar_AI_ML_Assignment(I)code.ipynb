```python
# @title categorizing
# 6. Adding HDI Category Column

# Define classification function based on UNDP thresholds
def classify_hdi(value):
    if value < 0.550:
        return "Low"
    elif 0.550 <= value <= 0.699:
        return "Medium"
    elif 0.700 <= value <= 0.799:
        return "High"
    else:
        return "Very High"

# Apply classification to create new column
hdi_2022_df.loc[:, 'HDI_Category'] = hdi_2022_df['hdi'].apply(classify_hdi)


# Verify classification
print(hdi_2022_df[['country', 'hdi', 'HDI_Category']].head())

# Save final dataframe with HDI category added
hdi_2022_df.to_csv("HDI_category_added.csv", index=False)
```

```python
# @title Problem 2: Advanced HDI Exploration

# Problem 2: Advanced HDI Exploration

# Objective: Perform advanced analysis of HDI data focusing on South Asian countries,
# composite metrics, outlier detection, metric relationships, and gap analysis.


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Task 1: Create South Asia Subset

# Define the list of South Asian countries as per assignment
south_asia_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India",
                        "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

# Filter dataset to include only South Asian countries
hdi_south_asia = df[df['country'].isin(south_asia_countries)].copy()

# Save the filtered dataset for submission
hdi_south_asia.to_csv("HDI_SouthAsia.csv", index=False)

print("South Asia subset created with shape:", hdi_south_asia.shape)



# Task 2: Composite Development Score

# Formula: Composite Score = 0.30 * Life Expectancy + 0.30 * GNI per Capita
# Note: We use raw columns 'life_expectancy' and 'gross_inc_percap' as proxies for indices.

hdi_south_asia['Composite_Score'] = (
    0.30 * hdi_south_asia['life_expectancy'] +
    0.30 * hdi_south_asia['gross_inc_percap']
)

# Rank countries by Composite Score (latest year 2022)
composite_rank = hdi_south_asia[hdi_south_asia['year'] == 2022].sort_values(
    by='Composite_Score', ascending=False)

print("\nComposite Score Ranking (2022):\n", composite_rank[['country','Composite_Score','hdi']])

# Plot top 5 countries in horizontal bar chart
plt.figure(figsize=(8,6))
sns.barplot(data=composite_rank.head(5), x="Composite_Score", y="country", palette="viridis")
plt.title("Top 5 South Asian Countries by Composite Score (2022)")
plt.xlabel("Composite Score")
plt.ylabel("Country")
plt.show()


# Task 3: Outlier Detection

# Outliers detected using 1.5 * IQR rule for HDI and GNI per Capita

def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return series[(series < lower) | (series > upper)]

# Detect outliers in HDI and GNI per Capita (2022 only)
hdi_outliers = detect_outliers(hdi_south_asia[hdi_south_asia['year']==2022]['hdi'])
gni_outliers = detect_outliers(hdi_south_asia[hdi_south_asia['year']==2022]['gross_inc_percap'])

print("\nHDI Outliers (2022):\n", hdi_outliers)
print("\nGNI per Capita Outliers (2022):\n", gni_outliers)

# Scatter plot highlighting outliers
plt.figure(figsize=(8,6))
subset_2022 = hdi_south_asia[hdi_south_asia['year']==2022]
sns.scatterplot(data=subset_2022, x="gross_inc_percap", y="hdi", hue="country", s=80)

# Highlight outliers in red
sns.scatterplot(data=subset_2022.loc[hdi_outliers.index], x="gross_inc_percap", y="hdi",
                color="red", s=120, label="HDI Outliers")
sns.scatterplot(data=subset_2022.loc[gni_outliers.index], x="gross_inc_percap", y="hdi",
                color="orange", s=120, label="GNI Outliers")

plt.title("Outlier Detection: GNI vs HDI (South Asia, 2022)")
plt.xlabel("Gross National Income per Capita")
plt.ylabel("HDI")
plt.legend()
plt.show()



# Task 4: Exploring Metric Relationships

# Select two metrics: Gender Development Index and Life Expectancy
metrics = ['gender_development', 'life_expectancy']

for metric in metrics:
    corr = subset_2022[metric].corr(subset_2022['hdi'])
    print(f"\nPearson correlation between {metric} and HDI (2022): {corr:.3f}")

    # Scatter plot with trendline
    plt.figure(figsize=(8,6))
    sns.scatterplot(data=subset_2022, x=metric, y="hdi", hue="country", s=80)
    sns.regplot(data=subset_2022, x=metric, y="hdi", scatter=False, color="red")
    plt.title(f"Relationship between {metric} and HDI (South Asia, 2022)")
    plt.xlabel(metric.replace("_"," ").title())
    plt.ylabel("HDI")
    plt.show()



# Task 5: Gap Analysis

# Define new metric: GNI_HDI_Gap = gross_inc_percap - hdi
subset_2022['GNI_HDI_Gap'] = subset_2022['gross_inc_percap'] - subset_2022['hdi']

# Rank countries by gap
gap_rank = subset_2022.sort_values(by='GNI_HDI_Gap', ascending=False)
print("\nTop Positive GNI-HDI Gaps:\n", gap_rank[['country','GNI_HDI_Gap']].head(3))
print("\nTop Negative GNI-HDI Gaps:\n", gap_rank[['country','GNI_HDI_Gap']].tail(3))

# Plot top 3 positive and negative gaps
plt.figure(figsize=(8,6))
sns.barplot(data=gap_rank.head(3), x="GNI_HDI_Gap", y="country", color="green", label="Positive Gap")
sns.barplot(data=gap_rank.tail(3), x="GNI_HDI_Gap", y="country", color="red", label="Negative Gap")
plt.title("Top 3 Positive & Negative GNI-HDI Gaps (South Asia, 2022)")
plt.xlabel("GNI - HDI Gap")
plt.ylabel("Country")
plt.legend()
plt.show()
```

```python

#@title Problem 3: Comparative Regional Analysis – South Asia vs Middle East

# Objective: Compare HDI and related metrics between South Asia and Middle East (2020–2022).


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Task 1: Create South Asia and Middle East Subsets

south_asia_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India",
                        "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

middle_east_countries = ["Bahrain", "Iran", "Iraq", "Israel", "Jordan",
                         "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
                         "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"]

# Filter dataset for 2020–2022 and copy safely
hdi_problem1B = df[df['year'].isin([2020, 2021, 2022])].copy()

# Create subsets and add region labels
south_asia_subset = hdi_problem1B[hdi_problem1B['country'].isin(south_asia_countries)].copy()
south_asia_subset['region'] = 'South Asia'

middle_east_subset = hdi_problem1B[hdi_problem1B['country'].isin(middle_east_countries)].copy()
middle_east_subset['region'] = 'Middle East'

# Save for submission
south_asia_subset.to_csv("HDI_SouthAsia_2020_2022.csv", index=False)
middle_east_subset.to_csv("HDI_MiddleEast_2020_2022.csv", index=False)

# Combine for comparison
combined_df = pd.concat([south_asia_subset, middle_east_subset])
print("Combined dataset shape:", combined_df.shape)


# Task 2: Descriptive Statistics

# Mean and standard deviation of HDI by region
region_stats = combined_df.groupby('region')['hdi'].agg(['mean','std']).reset_index()
print("\nDescriptive Statistics (HDI 2020–2022):\n", region_stats)



# Task 3: Top and Bottom Performers (2022)

subset_2022 = combined_df[combined_df['year'] == 2022]

top3_sa = subset_2022[subset_2022['region'] == 'South Asia'].nlargest(3, 'hdi')
bottom3_sa = subset_2022[subset_2022['region'] == 'South Asia'].nsmallest(3, 'hdi')

top3_me = subset_2022[subset_2022['region'] == 'Middle East'].nlargest(3, 'hdi')
bottom3_me = subset_2022[subset_2022['region'] == 'Middle East'].nsmallest(3, 'hdi')

# Combine and plot
top_bottom = pd.concat([top3_sa, bottom3_sa, top3_me, bottom3_me])

plt.figure(figsize=(10,6))
sns.barplot(data=top_bottom, x="hdi", y="country", hue="region")
plt.title("Top & Bottom HDI Performers (2022)")
plt.xlabel("HDI")
plt.ylabel("Country")
plt.show()


# Task 4: Metric Comparisons

metrics = ['gender_development', 'life_expectancy', 'gross_inc_percap']

for metric in metrics:
    plt.figure(figsize=(10,6))
    sns.barplot(data=subset_2022, x="region", y=metric, hue="country")
    plt.title(f"{metric.replace('_',' ').title()} Comparison by Region (2022)")
    plt.xlabel("Region")
    plt.ylabel(metric.replace("_"," ").title())
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()


# Task 5: HDI Disparity

# Range and coefficient of variation (CV = std/mean)
disparity = subset_2022.groupby('region')['hdi'].agg(['max','min','mean','std']).reset_index()
disparity['range'] = disparity['max'] - disparity['min']
disparity['cv'] = disparity['std'] / disparity['mean']
print("\nHDI Disparity by Region:\n", disparity[['region','range','cv']])


# Task 6: Correlation Analysis

for region in ['South Asia', 'Middle East']:
    region_data = subset_2022[subset_2022['region'] == region]
    corr_gender = region_data['gender_development'].corr(region_data['hdi'])
    corr_life = region_data['life_expectancy'].corr(region_data['hdi'])
    print(f"\n{region} Correlations (2022):")
    print("HDI vs Gender Development:", round(corr_gender,3))
    print("HDI vs Life Expectancy:", round(corr_life,3))

    # Scatter plots
    plt.figure(figsize=(8,6))
    sns.scatterplot(data=region_data, x="gender_development", y="hdi", hue="country", s=80)
    sns.regplot(data=region_data, x="gender_development", y="hdi", scatter=False, color="red")
    plt.title(f"{region}: HDI vs Gender Development (2022)")
    plt.show()

    plt.figure(figsize=(8,6))
    sns.scatterplot(data=region_data, x="life_expectancy", y="hdi", hue="country", s=80)
    sns.regplot(data=region_data, x="life_expectancy", y="hdi", scatter=False, color="red")
    plt.title(f"{region}: HDI vs Life Expectancy (2022)")
    plt.show()

# Task 7: Outlier Detection

def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return series[(series < lower) | (series > upper)]

for region in ['South Asia', 'Middle East']:
    region_data = subset_2022[subset_2022['region'] == region]
    hdi_outliers = detect_outliers(region_data['hdi'])
    gni_outliers = detect_outliers(region_data['gross_inc_percap'])

    print(f"\n{region} HDI Outliers:\n", region_data.loc[hdi_outliers.index][['country','hdi']])
    print(f"{region} GNI Outliers:\n", region_data.loc[gni_outliers.index][['country','gross_inc_percap']])

    # Scatter plot
    plt.figure(figsize=(8,6))
    sns.scatterplot(data=region_data, x="gross_inc_percap", y="hdi", hue="country", s=80)
    sns.scatterplot(data=region_data.loc[hdi_outliers.index], x="gross_inc_percap", y="hdi",
                    color="red", s=120, label="HDI Outliers")
    sns.scatterplot(data=region_data.loc[gni_outliers.index], x="gross_inc_percap", y="hdi",
                    color="orange", s=120, label="GNI Outliers")
    plt.title(f"{region}: Outlier Detection (2022)")
    plt.xlabel("Gross National Income per Capita")
    plt.ylabel("HDI")
    plt.legend()
    plt.show()
```
