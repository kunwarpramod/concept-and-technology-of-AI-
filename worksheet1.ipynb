```python
#problem 2 Array manipulation with NumPy
#1. Create an array with values ranging from 10 to 49
import numpy as np

array_10_to_49 = np.arange(10, 50)
print("Array from 10 to 49:\n", array_10_to_49)
```

```python
#7 reverse an array
arr = np.array([1, 2, 0, 0, 4, 0])
reversed_arr = arr[::-1]
print("Reversed array:\n", reversed_arr)
```

```python
#8 create a 2d array with 1 on border and 0 insice
border_array = np.ones((5, 5))
border_array[1:-1, 1:-1] = 0
print("Bordered array:\n", border_array)
```

```python
#9 create and 8x8 matrix with a checkerboard pattern
checkerboard = np.zeros((8, 8), dtype=int)
checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1
print("Checkerboard pattern:\n", checkerboard)
```

```python
#PROBLEM 3 : array operations with NumPy
import numpy as np

# Given arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])
```

```python
#1 add the two arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
add_xy = np.add(x, y)
print("x + y:\n", add_xy)
```

```python
#2 subtract the two arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
sub_xy = np.subtract(x, y)
print("x - y:\n", sub_xy)
```

```python
#3 multiply the array with any integer
x = np.array([[1, 2], [3, 5]])
mult_x = x * 3
print("x * 3:\n", mult_x)
```

```python
#4 find the square of each element of the array
x = np.array([[1, 2], [3, 5]])
square_x = np.square(x)
print("Square of x:\n", square_x)
```

```python
#5 dot product between v and w ,x and v , x and y
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])
dot_vw = np.dot(v, w)
print("v · w:", dot_vw)
dot_xv = np.dot(x, v)
print("x · v:\n", dot_xv)
dot_xy = np.dot(x, y)
print("x · y:\n", dot_xy)
```

```python
#6 concatenate x and y along row ; v and w along column
concat_xy_row = np.concatenate((x, y), axis=0)
print("Concatenate x and y along rows:\n", concat_xy_row)

concat_vw_col = np.vstack((v, w))
print("Concatenate v and w along columns:\n", concat_vw_col)
```

```python
#7 concatenate x and v; observe error
try:
    concat_xv = np.concatenate((x, v), axis=0)
    print("Concatenate x and v:\n", concat_xv)
except ValueError as e:
    print("Error:", e)
    print("Explanation: x is a 2D array (2x2), v is a 1D array (length 2). Shapes don't match for concatenation along axis 0.")
```

```python
#PROBLEM 4.2 HOW FAST IS NumPy
import numpy as np
import time
```

```python
#1 element - wise addition using python list
list1 = list(range(1000000))
list2 = list(range(1000000))

start = time.time()
result_list_add = [a + b for a, b in zip(list1, list2)]
end = time.time()
print("Python list addition time:", end - start, "seconds")

#using NumPy Arrays

array1 = np.arange(1000000)
array2 = np.arange(1000000)

start = time.time()
result_array_add = array1 + array2
end = time.time()
print("NumPy array addition time:", end - start, "seconds")
```

```python
#2 element-wise multiplication
#using python lists
start = time.time()
result_list_mul = [a * b for a, b in zip(list1, list2)]
end = time.time()
print("Python list multiplication time:", end - start, "seconds")
#using NumPy Arrays
start = time.time()
result_array_mul = array1 * array2
end = time.time()
print("NumPy array multiplication time:", end - start, "seconds")

```

```python
#3 Dot Product
#using python lists
start = time.time()
dot_list = sum(a * b for a, b in zip(list1, list2))
end = time.time()
print("Python list dot product time:", end - start, "seconds")

#using NumPy arrays
start = time.time()
dot_array = np.dot(array1, array2)
end = time.time()
print("NumPy array dot product time:", end - start, "seconds")

```

```python
#4 Matrix multiplication 1000x1000
#using NumPy lists
np_matrix1 = np.arange(1000000).reshape(1000, 1000)
np_matrix2 = np.arange(1000000).reshape(1000, 1000)

start = time.time()
result_matrix_np = np.dot(np_matrix1, np_matrix2)
end = time.time()
print("NumPy matrix multiplication time:", end - start, "seconds")


```
