```python
# @title
#problem 1
import pandas as pd

# Load the Titanic dataset
df = pd.read_csv('/content/drive/MyDrive/Concepts and technology of AI/week3/Titanic-Dataset (1).csv')  # Make sure 'titanic.csv' is in your working directory

# 1️ Create a DataFrame called 'fare' with only the Fare column
fare = df[['Fare']]
print("Fare column:")
print(fare.head())

# 2 Create a DataFrame called 'class_age' with Pclass and Age columns
class_age = df[['Pclass', 'Age']]
print("\nPclass and Age columns:")
print(class_age.head())

# 3 Create a DataFrame called 'survived_gender' with Survived and Sex columns
survived_gender = df[['Survived', 'Sex']]
print("\nSurvived and Sex columns:")
print(survived_gender.head())
```

```python
# @title
#problem 2 subsetting rows by categorical values

# 1 Filter passengers with fare > 100
fare_gt_100 = df[df['Fare'] > 100]
print("Passengers with fare > 100:")
print(fare_gt_100.head())

# 2 Filter passengers in first class (Pclass == 1)
first_class = df[df['Pclass'] == 1]
print("\nPassengers in first class:")
print(first_class.head())

# 3 Filter female passengers under age 18
female_under_18 = df[(df['Age'] < 18) & (df['Sex'] == 'female')]
print("\nFemale passengers under age 18:")
print(female_under_18.head())
```

```python
# @title
# 1 Filter passengers whose Embarked port is "C" or "S"
embarked_c_or_s = df[df['Embarked'].isin(['C', 'S'])]
print("Passengers who embarked from Cherbourg or Southampton:")
print(embarked_c_or_s.head())

# 2 Filter passengers whose Pclass is in [1, 2]
first_second_class = df[df['Pclass'].isin([1, 2])]
print("\nPassengers in first or second class:")
print(first_second_class.head())
```

```python
# @title
#exploratory data analysis practice 1
#  Handle missing Age values by filling with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# 1 Add fare_per_year column (Fare / Age)
df['fare_per_year'] = df['Fare'] / df['Age']

# 2 Subset rows where fare_per_year > 5
high_fare_age = df[df['fare_per_year'] > 5]

# 3Sort by descending fare_per_year
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)

# 4 Select Name and fare_per_year columns
result = high_fare_age_srt[['Name', 'fare_per_year']]

# 5 View the result
print("Passenger(s) with highest fare relative to age:")
print(result.head())
```

```python
# @title
# 1 Add fare_per_class column (Fare / Pclass)
df['fare_per_class'] = df['Fare'] / df['Pclass']

# 2Subset adult male passengers (Age ≥ 18 and Sex == 'male')
adult_males = df[(df['Age'] >= 18) & (df['Sex'] == 'male')]

# 3 Sort by descending fare_per_class
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)

# 4️Select Name, Age, and fare_per_class columns
result = adult_males_srt[['Name', 'Age', 'fare_per_class']]

# 5 View the result
print("Adult male passenger(s) with highest fare relative to class:")
print(result.head())
```
